# Linear Regression using California Housing Prices dataset
- Dataset from kaggle: https://www.kaggle.com/camnugent/california-housing-prices

```python
# import necessary libraries
import pandas as pd
import seaborn as sns
import sklearn as sk
from sklearn.model_selection import train_test_split
```

```python
# Uploads data and opens the first 5 rows
house_data = pd.read_csv("housing.csv")
house_data.head()
```

```python
# create a pairplot to visualise the data better
sns.pairplot(house_data)
plt.title("Housing Prices Pairplot")
```

```python
# split into features & label
X = house_data[['population']]
y = house_data[['total_rooms']]

# split into training and testing data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
lreg = LinearRegression()
lr = lreg.fit(X_train, y_train)
```

```python
# using the function "score" to see the accuracy of our model based on our test variables
print("Accuracy Score:", lr.score(X_test, y_test)*100)
```

## Plotting a Linear Regression Graph

```python
sns.regplot(X, y, ci=None)
```

## Extension: Multi-Variable Linear Regression

```python
# remove NaN values
house_data = house_data.dropna()
```

```python
# includes all columns except the "median_house_value" and "ocean_proximity" columns
X_multi = house_data.drop(["median_house_value", "ocean_proximity"], axis = 1)
y_multi = house_data[["median_house_value"]]
X_multi_train, X_multi_test, y_multi_train, y_multi_test = train_test_split(X_multi, y_multi, test_size = 0.3, random_state = 0)
```

```python
# train model
lr_multi = lreg.fit(X_multi_train, y_multi_train)
# get the accuracy
print("Accuracy Score:", lr_multi.score(X_multi_test, y_multi_test)*100)
```
